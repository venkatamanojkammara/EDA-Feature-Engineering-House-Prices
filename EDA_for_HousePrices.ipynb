Import the required and necessary python libraries to perform the Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

Read the Dataset and viualise the top 5 rows of the dataset

```python
dataset = pd.read_csv('train.csv')
dataset.head()
```

Now here we have the find, if there were missing values with respect to the feature. if there were any missing values in them there we have to find the percentage of the missing values in that specific feature.

```python
features_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
print(features_with_nan)
```

```python
for feature in features_with_nan:
  print(feature,':',np.round(dataset[feature].isnull().mean(),4),'% of missing values')
```

we also need to find the no.of missing values in each feature


```python
for feature in features_with_nan:
  print(feature,':',dataset[feature].isnull().sum(),'missing values')
```

here form the above two blocks of code we can clearly observe that there are the missing values in the data set. now it is necessarhy that we need to see if there was any dependency of the missing values with the output in this case (SalePrice).

We see the above observations by plotting the bar graphs

```python
for feature in features_with_nan:
  data=dataset.copy()
  data[feature]=np.where(data[feature].isnull(),1,0)
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel("SalePrice")
  plt.title(feature)
  plt.show()
```

From the above graphs we observe that, in some graphs the 'one' bar is bigger than the 'zero' bar. in that case then there must be an effect of those missing values to the output 'SalePrice'

Upto here one step was completed and now it turn to check for the numerical variables and theie dependency with the output

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']
print(numerical_features)
print("Total no.of numerical features are ", len(numerical_features))
```

```python
dataset[numerical_features].head()
```

here from the above dataset we have data-time variables ie. year features. There this type of features are dependent on the past data.
Now we will find the list of features that contain the year feature.


```python
year_feature=[feature for feature in dataset.columns if 'Yr' in feature or 'Year' in feature]
print(year_feature)
```

now we will see the year values for corresponding to the year features.

```python
for feature in year_feature:
  print(feature, dataset[feature].unique())
```

Now we consider one paramenter and we compare or check the dependency on the output SalePrice by plotting a graph


```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel("SalePrice")
plt.title("YrSold")
plt.show()
```

here from the above graph we can clearly observe that. as the incerases then the sales price gradually decreases.

similar to the above case we will comare or observe for all remaining year features.

```python
for feature in year_feature:
  if feature != 'YrSold':
    data=dataset.copy()

    data[feature]=data['YrSold']-data[feature]
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
for feature in year_feature:
  if feature != 'YrSold':
    data=dataset.copy()


    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

Here we have two types of numerical features in the above data sets.
1. continuous variables
2. discrete variables

Now we will find the discrete variables of considering length less than 25 and they should not present in the year_features and 'Id'

```python
discrete_features=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature + ['Id']]
print(discrete_features)
print(len(discrete_features))
```

```python
dataset[discrete_features].head()
```

Now lets find the relation between the discrete features and the output saleprice but plotting some graphs

```python
for feature in discrete_features:
  data=dataset.copy()

  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
```

```python
for feature in discrete_features:
  print(feature,dataset[feature].unique(),len(dataset[feature].unique()))
```

according to the above information the bar graphs were plotted between the discrete features and the saleprice

now we will see the categorical features


```python
continuous_features=[feature for feature in numerical_features if feature not in discrete_features + year_feature+['Id']]
print(continuous_features)
print(len(continuous_features))
```

now we will check the relation between the continuous features and the saleprice by plotting some histograms

```python
for feature in continuous_features:
  data=dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel('count')
  plt.title(feature)
  plt.show()
```

Now we will apply the logarithumic transformation to the continuous feature values


```python
for feature in continuous_features:
  data=dataset.copy()
  if 0 in dataset[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('sale price')
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_features:
  data=dataset.copy()
  if 0 in dataset[feature].unique():
    print("0 in dataset[{}] {}".format(feature, len(feature)))
  else:
    print('0 not in dataset[{}] {}'.format(feature,len(data)))
```

Now we will find the outliers, as these otliers plays some impact on the output. Outliers are the values of the features that are extremely maximum or extremely minimum.
We can find, if there were any outliers in the feature by plotting the box plot.

```python
for feature in continuous_features:
  data=dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

From the above figures some values are present at extreme positions, ie they are the outliers.
